```python
import numpy as np
import pandas as pd
import tensorflow as ptf
from tensorflow import keras
import matplotlib.pyplot as plt
import os
from sklearn import preprocessing 
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer,MinMaxScaler,StandardScaler
```

```python
path = 'D:\ml\datasets\hr'
train = pd.read_csv(os.path.join(path,'train_LZdllcl.csv'))
test = pd.read_csv(os.path.join(path,'test_2umaH9m.csv'))
```

```python
#one hot encoding for department,region,education,recruitment
department = list()
for dep in train['department']:
    if dep=='Analytics':
        department.append(1)
    elif dep == 'Finance':
        department.append(2)
    elif dep == 'HR':
        department.append(3)
    elif dep == 'Legal':
        department.append(4)
    elif dep == 'Operations':
        department.append(5)
    elif dep == 'Procurement':
        department.append(6)
    elif dep == 'R&D':
        department.append(7)
    elif dep == 'Sales & Marketing':
        department.append(8)
    elif dep == 'Technology':
        department.append(9)
        
        
region_first = list()
for r in train['region']:
    region_first.append(int(r[7:]))
    
df_region = pd.DataFrame(region_first)
norm = MinMaxScaler()
df_scaled = norm.fit_transform(df_region)
region = df_scaled[:,0].tolist()    

    
    
education = list()
for edu in train['education']:
    if edu == "Bachelor's":
        education.append(1)
    elif edu == "Below Secondary":
        education.append(2)
    elif edu == "Master's & above":
        education.append(3)
    else:
        education.append(0)
        
        
        
        
gender = list()
for gen in train['gender']:
    if gen == 'm':
        gender.append(0)
    else:
        gender.append(1)
        
        
        
        
rec_channel = list()
for rec in train['recruitment_channel']:
    if rec == "referred":
        rec_channel.append(1)
    elif rec == "sourcing":
        rec_channel.append(2)
    else:
        rec_channel.append(0)
```

```python
my_train = train.copy()

my_train['department'] = department
df_dep = pd.DataFrame(my_train['department'])
norm = MinMaxScaler()
df_fit = norm.fit(df_dep)
df_scaled = df_fit.transform(df_dep)
department = df_scaled[:,0].tolist() 

df_age = pd.DataFrame(my_train['age'])
norm = MinMaxScaler()
df_fit = norm.fit(df_age)
df_scaled = df_fit.transform(df_age)
age = df_scaled[:,0].tolist()    


df_rate = pd.DataFrame(my_train['previous_year_rating'])
norm = MinMaxScaler()
df_fit = norm.fit(df_rate)
df_scaled = df_fit.transform(df_rate)
rate = df_scaled[:,0].tolist()    


df_service = pd.DataFrame(my_train['length_of_service'])
norm = MinMaxScaler()
df_fit = norm.fit(df_service)
df_scaled = df_fit.transform(df_service)
service = df_scaled[:,0].tolist()    


df_score = pd.DataFrame(my_train['avg_training_score'])
norm = MinMaxScaler()
df_fit = norm.fit(df_score)
df_scaled = df_fit.transform(df_score)
score = df_scaled[:,0].tolist()    
```

```python
my_train['department'] = department
my_train['education'] = education
my_train['gender'] = gender
my_train['recruitment_channel'] = rec_channel
my_train['region'] = region
my_train['avg_training_score'] = score
my_train['length_of_service'] = service
my_train['previous_year_rating'] = rate
my_train['age'] = age
```

```python
my_train_na = my_train.fillna(0)
my_train_drop = my_train_na.drop(columns=['department','employee_id','recruitment_channel','gender'])

X = my_train_na.iloc[:,0:-1]
Y = my_train_na.iloc[:,-1]

# x_train,x_val,y_train,y_val = train_test_split(X,Y,test_size=0.25,random_state=7)
```
